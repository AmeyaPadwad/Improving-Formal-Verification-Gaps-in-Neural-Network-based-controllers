```python
from neural_nets import PolicyNetwork
from verifiers import verifier
from systems import InvertedPendulum, CartPole
import numpy as np
```

```python
pendulum = InvertedPendulum()
network = PolicyNetwork(pendulum, device='cpu')
network.load_model("./nnets/pend_base.pth")
```

```python
v = verifier(pendulum, network)
n_samples = 1000

nominal_state = np.array([0.0, 0.0])
nominal_output = network(nominal_state)

output_range_sampling, output_samples = v.output_range_sampling(n_samples)
output_range_ibp, time_elapsed_ibp = v.interval_bound_propagation()

output_range_ibp_auto, time_elapsed_ibp_auto = v.auto_lirpa("ibp")
output_range_crown_auto, time_elapsed_crown_auto = v.auto_lirpa("crown")
output_range_alpha_crown_auto, time_elapsed_alpha_crown_auto = v.auto_lirpa("alpha-crown")


print(f"{nominal_output=}")
print(f"{output_range_sampling=}")
print(f"{output_range_ibp=}, {time_elapsed_ibp=}s")
print(f"{output_range_ibp_auto=}, {time_elapsed_ibp_auto=}s")
print(f"{output_range_crown_auto=}, {time_elapsed_crown_auto=}s")
print(f"{output_range_alpha_crown_auto=}, {time_elapsed_alpha_crown_auto=}s")
```
