```python
ac_cart = ActorCriticController(cartpole, actor_lr=3e-4, critic_lr=1e-3)
ac_cart.train(num_episodes=1000, eval_interval=50)

print("\nCartPole Deterministic Tests:")
test_cases_cart = [
    (np.array([0.0, 0.0, 0.05, 0.0]), "Small angle (0.05)"),
    (np.array([0.0, 0.0, 0.15, 0.0]), "Medium angle (0.15)"),
    (np.array([0.0, 0.0, 0.2, 0.0]), "Large angle (0.2)"),
    (np.array([1.5, 0.0, 0.0, 0.0]), "Cart position (1.5)"),
    (np.array([1.0, 0.0, 0.1, 0.0]), "Combined (1.0, 0.1)"),
    (np.array([0.5, 1.0, 0.1, 0.5]), "With velocities (0.5, 1.0, 0.1, 0.5)"),
    (np.array([0.0, 0.0, -0.15, 0.0]), "Negative angle (-0.15)"),
    (np.array([-1.5, 0.0, 0.0, 0.0]), "Negative position (-1.5)"),
]

for initial_state, label in test_cases_cart:
    success, final, traj = ac_cart.test_single_trajectory(initial_state, verbose=False)
    print(f"  {label}: {'✓ PASS' if success else '✗ FAIL'}")

success_rate = ac_cart.evaluate(num_trials=500, max_steps=200)
print(f"\nCartPole AC Random Trials: {success_rate:.1f}% success rate")
```
