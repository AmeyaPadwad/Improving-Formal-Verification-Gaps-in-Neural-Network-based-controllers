```python
def plot_accs(accs):
    l = accs[1:]
    acc_diffs = [i[1]-i[0] for i in l]
    mag = acc_diffs[:4]
    act = acc_diffs[4:]
    
    plt.plot(mag, marker='o', label="Magnitude pruning")
    plt.plot(act, marker='o', label="Activation pruning")
    plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
    plt.title("Change in accuracy with pruning")
    plt.xticks([0, 1, 2, 3], [0.1, 0.3, 0.5, 0.7])
    plt.legend()
    plt.show()

plot_accs(pendulum_relu_accuracies)
```

```python
def plot_output_bound_widths(bounds_lists, titles):
    fig, ax = plt.subplots(figsize=(14, 6))
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    bound_widths = []
    for sys in bounds_lists:
        width = []
        for bounds in sys:
            width.append(abs(bounds[0]-bounds[1]))
        bound_widths.append(width)
    
    bound_widths_mag = []
    for bounds in bound_widths:
        bound_widths_mag.append(bounds[:5])
    
    bound_widths_act = []
    for bounds in bound_widths:
        bound_widths_act.append([bounds[0]]+bounds[5:])

    for i, w in enumerate(bound_widths_mag):
        ax.plot(w, marker='o', label=titles[i], color = colors[i])
        ax.plot(bound_widths_act[i], marker='o', linestyle ='--', color = colors[i])


    plt.title("Change in bound witdth with pruning - cartpole relu")
    plt.xticks([0, 1, 2, 3, 4], [0, 0.1, 0.3, 0.5, 0.7])
    plt.xlabel("Pruning Sparsity")
    plt.ylabel("Bound width")
    plt.legend()
    plt.yscale("log")
    plt.show()

bounds_lists = [
    cartpole_relu_sampling_output_bounds,
    cartpole_relu_ibp_output_bounds,
    cartpole_relu_crown_output_bounds,
    cartpole_relu_alpha_crown_output_bounds,
    cartpole_relu_lipschitz_output_bounds
]

titles = ['Sampling', 'IBP', 'CROWN', 'Alpha-CROWN', 'Lipschitz']

plot_output_bound_widths(bounds_lists, titles)
```

```python
def plot_output_bounds(bounds_lists, titles):
    """
    Plot output bounds for different verification methods on the same plot.
    
    Args:
        bounds_lists: List of bounds lists (e.g., [sampling, ibp, crown, alpha_crown, lipschitz])
        titles: List of titles corresponding to each bounds list
    """
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # Colors for each method
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
    
    # Create positions for each configuration
    positions = np.arange(9)
    spacing = 0.12
    
    for method_idx, (bounds, title, color) in enumerate(zip(bounds_lists, titles, colors)):
        # Extract lower and upper bounds
        lower_bounds = [b[0] for b in bounds]
        upper_bounds = [b[1] for b in bounds]
        
        # Offset positions for each method
        offset_positions = positions + (method_idx - 2) * spacing
        
        # Plot as vertical lines showing the bounds (lower to upper)
        for pos, lower, upper in zip(offset_positions, lower_bounds, upper_bounds):
            ax.plot([pos, pos], [lower, upper], color=color, linewidth=2, alpha=0.7)
            ax.plot(pos, lower, 'o', color=color, markersize=4)
            ax.plot(pos, upper, 'o', color=color, markersize=4)
    
    # Add vertical lines to separate base from magnitude and activation pruning
    ax.axvline(x=0.5, color='gray', linestyle='--', alpha=0.3)
    ax.axvline(x=4.5, color='gray', linestyle='--', alpha=0.3)
    
    labels = ['Base', 'Mag 0.1', 'Mag 0.3', 'Mag 0.5', 'Mag 0.7', 
              'Act 0.1', 'Act 0.3', 'Act 0.5', 'Act 0.7']
    ax.set_xticks(positions)
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.set_ylabel('Output Bounds')
    ax.set_title('Output Bounds Across Verification Methods')
    ax.grid(axis='y', alpha=0.3)
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.legend(titles, loc='best')
    
    plt.tight_layout()
    plt.show()


bounds_lists = [
    pendulum_relu_sampling_output_bounds,
    pendulum_relu_ibp_output_bounds,
    pendulum_relu_crown_output_bounds,
    pendulum_tanh_alpha_crown_output_bounds,
    pendulum_tanh_lipschitz_output_bounds
]

titles = ['Sampling', 'IBP', 'CROWN', 'Alpha-CROWN', 'Lipschitz']

plot_output_bounds(bounds_lists, titles)
```
